```python
import geopandas as gpd
import folium as fm
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# Construire un DataFrame pour chaque fichier CSV
df_c = pd.read_csv("data/caract-2024.csv", on_bad_lines='skip', delimiter=';')
df_l = pd.read_csv("data/lieux-2024.csv", on_bad_lines='skip', delimiter=';')
df_u = pd.read_csv("data/usagers-2024.csv", on_bad_lines='skip', delimiter=';')
df_v = pd.read_csv("data/vehicules-2024.csv", on_bad_lines='skip', delimiter=';')
```

```python
df_v.drop(columns=['id_vehicule\n'], inplace=True)
```

```python
# Joindre les DataFrames en utilisant Num_Acc comme clé
df = df_c.merge(df_l, on='Num_Acc', how='inner') \
                .merge(df_u, on='Num_Acc', how='inner') \
                .merge(df_v, on='Num_Acc', how='inner')

# Afficher les premières lignes du DataFrame fusionné
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
df.drop(columns=['lartpc'], inplace=True)
```

```python
# Remplacer les valeurs nulles de 'an_nais' par sa médiane
df['an_nais'].fillna(df['an_nais'].median(), inplace=True)

# Remplacer les valeurs nulles de 'occutc' par sa médiane
df['occutc'].fillna(df['occutc'].median(), inplace=True)
```

```python
df.dropna(subset=['adr', 'voie'], inplace=True)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# Calculer l'âge au moment de l'accident
df['age'] = df['an'] - df['an_nais']

# Supprimer la colonne 'an_nais'
df.drop(columns=['an_nais'], inplace=True)

# Afficher les premières lignes pour vérifier
print(df[['an', 'age']].head())
```

```python
df.to_csv("data/df_dataset.csv", index=False)
```

```python
# Filtrer les données pour les blessés graves
blessures_graves = df[df['grav'] == 2]

# Compter le nombre de blessés graves par département
departements_blessures = blessures_graves['dep'].value_counts()

# Afficher toutes les colonnes
pd.set_option('display.max_columns', None)

# Afficher toutes les lignes
pd.set_option('display.max_rows', None)
# Afficher les départements avec le plus de blessés graves
print(departements_blessures)
```
